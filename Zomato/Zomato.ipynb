#  Zomato Dataset: Exploratory Data Analysis

### EDA involves:

1. Understand the structure, patterns, and relationships

2. Summarize main characteristics often using visualization

3. Identify missing values, duplicates, or outliers

#### In this notebook, I performed EDA using:

.info(), .describe(), .isnull().sum() for data overview

sns.countplot(), sns.barplot(), and df.value_counts() for understanding distributions

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
%matplotlib inline
```

```python
# Importing the dataset
df=pd.read_csv('zomato.csv',encoding='latin-1') # encoding to handle character encoding when dealing with non-UTF-8 encoded files
df.head(5)
```

```python
# columns in the dataset
df.columns
```

```python
# understanding the dataset
df.info()
```

```python
# Descriptive statistics of the data ( only numerical values)
df.describe()
```

## Data Analysis
1. Missing values
2. Explore numerical variables
3. Explore categorical variables
4. Find relationship between features

```python
df.shape
```

```python
# How many null values are there
df.isnull().sum()  
```

```python
# which specific feature has null values
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
sns.heatmap(df.isnull(),yticklabels=False, cbar=False,cmap='viridis')
matplotlib.rcParams['figure.figsize']=(12,6) # Increase the figsize
```

```python
# Importing of the excel file (Country-Code.xlsx)
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# Combine the two datasets (Zomato csv & country code xlsx)
final_df =pd.merge(df,df_country, on ='Country Code', how='left') # Country code is the point of merge
```

```python
final_df
```

```python
# to check the data types
final_df.dtypes
```

```python
final_df.columns
```

```python
# how many records are there with the specific country
final_df.Country.value_counts()
```

### Observations:
8,652 records are from India, which is overwhelmingly more than any other country.
This suggests the Zomato dataset is India-centric, possibly because Zomato originated and operates primarily in India.

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# Pie-Chart - Top 3 Countries that use Zomato
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%') #to add percentage to the chart
```

### Observations:
1. India dominates the dataset with 94.39% of records.

2. United States and United Kingdom follow with 4.73% and 0.87%, respectively.

```python
final_df.columns
```

```python
# Groups the dataset by aggregate rating, rating color, and rating text.
# Counts the number of records in each group.
# Converts the group results into a DataFrame 
#  Renames the auto-generated column 
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})  
```

```python
ratings
```

### Observations:
1. Ratings between 4.5 to 4.9 - Excellent
2. Ratings between 4.0 to 4.4 - very Good
3. Ratings between 3.5 to 3.9 - Good
4. Ratings between 2.5 to 3.4 - Average
5. Ratings between 1.8 to 2.4 - Poor

```python
ratings.head()
```

```python
# Sets the plot size to make the chart wider and easier to read.
matplotlib.rcParams['figure.figsize']=(12,6)
# bar plot showing the count of records for each aggregate rating.
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

### Observations:
1. The chart highlights that a significant portion of restaurants are unrated
2. Among rated restaurants, most receive mid-level ratings, suggesting average performance is common, and extreme ratings are rare.

```python
ratings
```

```python
# bar plot showing the count of restaurants in each rating color category based on ratings
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
# Find the country name that has given 0 rating
final_df[final_df['Rating color']=='White'].groupby('Country').size().reset_index()
```

### Observations:
1. max no of 0 ratings is from India then Brazil,US,UK

```python
# Find which currency is used by each country
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```
